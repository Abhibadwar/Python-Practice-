```python
# find factorial(loop)
n = 5
fact = 1
for i in range(1, n+1):
    fact *= i
print(fact)
```

```python
# sum of firstN number 
n = 10 
total = n * (n+1) // 2
print(total)
```

```python
# reverse a string
a = "Abhishek"
print(a[::-1])
```

```python
s = 'sir'
if s == s[::-1]:
    print("palindrome")
else:
    print("Not")
```

```python
# fibonacci series
a,b = 0,1
for _ in range(10):
    print(a, end=" ")
    a, b = b, a + b
```

```python
# find max of three number 
a,b,c = 3,7,5
print(max(a,b,c))
```

```python
# check prime number 
n = 20
prime = True 
for i in range(2, int(n**0.5) + 1):
    if n % i == 0:
        prime = False 
        break

print("prime"if prime else "not Prime")
```

```python
# find sum of digits
n = 1234
print(sum(int(d)for d in str(n)))
```

```python
# count vowels
a = 'abhishek'
vowels = "aeiou"
count = sum(1 for i in a if i in vowels)
print(count)
```

```python
# multiplication table of a number 
n = 10
for i in range(1,11):
    print(n,"x",i,"=",n*i)
```

```python
# check leap year)
y = 2028
if (y % 4 == 0 and y % 100 != 0) or (y % 400 == 0):
    print('yes')
else:
    print('no')
```
