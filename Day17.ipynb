```python
# remove dulicate
def remove_dulicate(n):
    o = []
    for i in n:
        if i not in o:
            o. append(i)
    return o
```

```python
n = [1,1,2,2,3,3,45,5,6,8,8]
remove_dulicate(n)
```

```python
# reverse list
def reverse(s):
    reversed =""
    for i in s:
        reversed = i + reversed
    return reversed 
```

```python
s = "Hello world"
reverse(s)
```

```python
#  Write a promgram for below input 
# input "aceg"
# out -='bdfh'

def shift_s(a):
    s =""
    for i in a:
        s += chr(ord(i) + 1)
    return s
```

```python
a = "aceg"
shift_s(a)
```

```python
# implement a function to check if a given string is a palindrome
def palindrome(a):
    a = "".join([i.lower() for i in a if i .isalnum()])
    return a == a[::-1]
```

```python
a = " abhishe badwar i am form nashik"
palindrome(a)
```

```python
# write a program to find the factorial of a number recursion
def fac(a):
    if a == 0 or a == 1:
        return 1
    else:
        return a * fac(a-1)
```

```python
fac(5)
```

```python
#  Given a list of integers write a function to return the sum of all postitive number in the lits 
def sum_positive(n):
    return sum(num for num in n if num > 0)
```

```python
# write prime number 
def prime(n):
    if n <= 1:
        return False
    for i in range (2 ,int(n ** 0.5) + 1):
            if  n % i == 0:
                return False 
            return True
```

```python
prime(23)
```

```python
# program to find pairs in array whose sum matches target
def find_pairs(a,t):
    p = []
    s = set()
    for n in a:
        com = t - n
        if com in s:
            p.append((com, n))
        s.add(n)
    return p
```

```python
a = [1,2,3,4,5,6,7,8]
t = 10
find_pairs(a,t)
```
