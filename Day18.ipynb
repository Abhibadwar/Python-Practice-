```python
#Write a Python function to print the first n terms of the Fibonacci sequence.
def fibo_sequence(n):
    list = []
    for n in range(0,n):
        if n == 0 or n == 1:
            list.append(n)
        else:
            list.append(list[-1] +list[-2])
    return list

fibo_sequence(1)
    
```

```python
def fibo(n):
    fib = [0,1]
    while len(fib) < n:
        fib.append(fib[-1]+fib[-2])
    return fib[:n]
fibo(5)  
```

```python
def fib(n):
    if n<=0:
        return 0
    elif n == 1:
        return 1
    else:
        return fib(n-1) + fib(n-1)
```

```python
fib(5)
```

```python
## Write a Python program that counts the number of vowels and consonants in a given string.
s = "abhishek"
vowels = "aeiouAEIOU"
v_count, c_count = 0, 0

for ch in s:
    if ch in vowels:
        v_count += 1
    else:
        c_count += 1

print("Vowels:", v_count, "Consonants:", c_count)
```

```python
## Write two Python functions to calculate the factorial of a number—one using recursion and the other using a loop.
def fact_recurcive(n):
    if n == 0 or n ==1:
        return 1
    return n * fact_recurcive(n-1)
fact_recurcive(5)
```

```python
def fact_iterative(n):
    r = 1
    for i in range(2,n+1):
        r *= i
    return r
fact_iterative(4)
```

```python
def fact(n):
    if n == 0:
        return 1
    return n *fact(n-1)
fact(4)
```
