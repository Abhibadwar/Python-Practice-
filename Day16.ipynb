```python
# counting special char in string 
def count(s):
    sp = "!@#$%^&*(){}[];:'""~`,.<>"
    c = 0
    for i in s:
        if i in sp:
            c+=1
    return c
```

```python
t = "Hello! Abhishek # how are doing $ i am * fine () What about> you"
r = count(t)
r
```

```python
def replace(s):
    s = s.replace(',','third')
    s = s.replace('.','.')
    s = s.replace('third',',')
    return s
    
```
